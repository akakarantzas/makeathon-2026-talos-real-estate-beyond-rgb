# Spectral Analysis of Candidate Plots

This notebook loads each plot's `SPECTRAL_IMAGE.TIF`, reports basic raster metadata, computes vegetation and moisture indices, and plots the results side by side for comparison.

Expected data layout:
```text
data/
├── Arkadia/SPECTRAL_IMAGE.TIF
├── Magnisia/SPECTRAL_IMAGE.TIF
├── Arkadia_2/SPECTRAL_IMAGE.TIF
└── Veroia/SPECTRAL_IMAGE.TIF
```


````python
from pathlib import Path
import re

import matplotlib.pyplot as plt
import numpy as np
import rasterio
````

## Discover plot imagery

The notebook searches recursively under `data/` for files named `SPECTRAL_IMAGE.TIF`. It expects exactly four images, one per plot.


````python
DATA_DIR = Path('../data')
image_paths = sorted(DATA_DIR.rglob('SPECTRAL_IMAGE.TIF'))

if len(image_paths) != 4:
    raise FileNotFoundError(
        f'Expected 4 SPECTRAL_IMAGE.TIF files under {DATA_DIR.resolve()}, found {len(image_paths)}: {image_paths}'
    )

plot_paths = {path.parent.name: path for path in image_paths}
plot_paths
````

## Band selection

For hyperspectral imagery, the exact band number depends on the product. The code below first tries to extract wavelengths from raster metadata or band descriptions and chooses the nearest bands to common targets:

- Green: 560 nm
- Red: 665 nm
- NIR: 842 nm
- SWIR: 1640 nm

If your TIFF files do not expose wavelength metadata, fill in `MANUAL_BAND_INDEXES` with 1-based raster band numbers before running the analysis cells.


````python
TARGET_WAVELENGTHS_NM = {
    'green': 560,
    'red': 665,
    'nir': 842,
    'swir': 1640,
}

# Optional fallback if wavelength metadata is unavailable.
# Use 1-based raster band numbers, e.g. {'green': 42, 'red': 58, 'nir': 91, 'swir': 170}
MANUAL_BAND_INDEXES = None

def extract_wavelengths_nm(src):
    candidates = []

    for index in range(1, src.count + 1):
        values = []
        description = src.descriptions[index - 1]
        if description:
            values.append(description)
        values.extend(src.tags(index).values())

        wavelength = None
        for value in values:
            match = re.search(r'(\d+(?:\.\d+)?)\s*(?:nm|nanometer)', str(value), re.IGNORECASE)
            if match:
                wavelength = float(match.group(1))
                break
        candidates.append(wavelength)

    if any(value is not None for value in candidates):
        return np.array(candidates, dtype=float)

    return None

def select_band_indexes(src):
    if MANUAL_BAND_INDEXES is not None:
        return MANUAL_BAND_INDEXES

    wavelengths = extract_wavelengths_nm(src)
    if wavelengths is None:
        raise ValueError(
            'No wavelength metadata found. Set MANUAL_BAND_INDEXES to 1-based band numbers before continuing.'
        )

    selected = {}
    for name, target_nm in TARGET_WAVELENGTHS_NM.items():
        selected[name] = int(np.nanargmin(np.abs(wavelengths - target_nm))) + 1
    return selected
````

## Load metadata and compute indices

- **NDVI** = `(NIR - Red) / (NIR + Red)`
- **NDWI / moisture index** here uses the Gao formulation: `(NIR - SWIR) / (NIR + SWIR)`


````python
def safe_normalized_difference(a, b):
    a = a.astype('float32')
    b = b.astype('float32')
    denominator = a + b
    return np.divide(a - b, denominator, out=np.full_like(a, np.nan, dtype='float32'), where=denominator != 0)

results = {}

for plot_name, path in plot_paths.items():
    with rasterio.open(path) as src:
        band_indexes = select_band_indexes(src)
        green = src.read(band_indexes['green'])
        red = src.read(band_indexes['red'])
        nir = src.read(band_indexes['nir'])
        swir = src.read(band_indexes['swir'])

        ndvi = safe_normalized_difference(nir, red)
        ndwi = safe_normalized_difference(nir, swir)

        results[plot_name] = {
            'path': path,
            'bands': src.count,
            'resolution': src.res,
            'crs': src.crs,
            'band_indexes': band_indexes,
            'ndvi': ndvi,
            'ndwi': ndwi,
        }

        print(f'{plot_name}:')
        print(f'  file: {path}')
        print(f'  bands: {src.count}')
        print(f'  resolution: {src.res}')
        print(f'  CRS: {src.crs}')
        print(f'  selected bands: {band_indexes}')
        print()
````

## Compare plots side by side


````python
plot_names = list(results.keys())
fig, axes = plt.subplots(2, len(plot_names), figsize=(5 * len(plot_names), 9), constrained_layout=True)

for col, plot_name in enumerate(plot_names):
    ndvi = results[plot_name]['ndvi']
    ndwi = results[plot_name]['ndwi']

    im0 = axes[0, col].imshow(ndvi, cmap='RdYlGn', vmin=-1, vmax=1)
    axes[0, col].set_title(f'{plot_name} - NDVI')
    axes[0, col].axis('off')

    im1 = axes[1, col].imshow(ndwi, cmap='BrBG', vmin=-1, vmax=1)
    axes[1, col].set_title(f'{plot_name} - NDWI')
    axes[1, col].axis('off')

fig.colorbar(im0, ax=axes[0, :], shrink=0.8, label='NDVI')
fig.colorbar(im1, ax=axes[1, :], shrink=0.8, label='NDWI / moisture index')
plt.show()
````
